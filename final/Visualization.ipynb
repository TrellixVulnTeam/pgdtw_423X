```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
_id = 5426
data = pd.read_csv('cv/%s.csv'%_id, header=None, index_col=None)
data.columns = ['batch_id', 't', 'T', 'est_T']
data.sort_values(by=['t'], inplace=True)
true_len = data['T'][0]
dev = data['T'].values - (data['t'].values + data['est_T'].values)
plt.plot(data['t'].values, dev)
plt.plot([0, true_len], [true_len, 0])
plt.plot([0, true_len], [-true_len, 0])
plt.ylim((min(dev)-10, max(dev)+10))
plt.show()
```

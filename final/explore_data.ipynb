```python
import libdtw as lib
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = lib.load_data(1000)
len(data['5153'])
```

```python
D = lib.Dtw(data)
```

```python
len(D.data['reference'])
```

```python
lengths = list()
for q in D.data['queries'].values():
    lengths.append(len(q[0]['values']))
print('Min: %d \nMax: %d \nMean: %0.1f \nMedian: %d' %(min(lengths), max(lengths), np.mean(lengths), np.median(lengths)))
```

```python
plt.hist(lengths, color='#d90000')
plt.ylabel('Count')
plt.xlabel('Duration (min)')
plt.show()
```

```python
plt.hist(lengths, color='#787878', bins = 20)
plt.ylabel('Count')
plt.xlabel('Duration (min)')
plt.vlines(x = 415, ymin =0, ymax = 73, colors='#000000', label='Reference batch')
plt.legend()
plt.savefig('images_thesis/queries_len_distro.png')
plt.show()
```

```python
# correlation length batchID
import datetime
from matplotlib.dates import date2num
fig = plt.figure(figsize=(15,5))
for _id, q in D.data['queries'].items():
    plt.plot_date(date2num(datetime.datetime.strptime(q[0]['start'], '%Y-%m-%d %H:%M:%S')), len(q[0]['values']), color = '#000000')

plt.axvspan(date2num(datetime.datetime.strptime('2014-11-01', '%Y-%m-%d')), 
            date2num(datetime.datetime.strptime('2014-12-31', '%Y-%m-%d')), 
            color='#787878', alpha=0.1)

plt.axvspan(date2num(datetime.datetime.strptime('2015-01-01', '%Y-%m-%d')), 
            date2num(datetime.datetime.strptime('2015-12-31', '%Y-%m-%d')), 
            color='#787878', alpha=0.5)

plt.axvspan(date2num(datetime.datetime.strptime('2016-01-01', '%Y-%m-%d')), 
            date2num(datetime.datetime.strptime('2016-12-31', '%Y-%m-%d')), 
            color='#787878', alpha=0.1)

plt.axvspan(date2num(datetime.datetime.strptime('2017-01-01', '%Y-%m-%d')), 
            date2num(datetime.datetime.strptime('2017-12-31', '%Y-%m-%d')), 
            color='#787878', alpha=0.5)

plt.xlabel('Start date', fontsize = 20)
plt.xticks(fontsize=20)
plt.ylabel('Batch duration (min)', fontsize=20)
plt.yticks(fontsize=20)
plt.xlim((date2num(datetime.datetime.strptime('2014-11-01', '%Y-%m-%d')), 
          date2num(datetime.datetime.strptime('2017-12-31', '%Y-%m-%d'))))
plt.tight_layout()
plt.savefig('C:\\Users\\DEPAGRA\\Dropbox\\Applicazioni\\Overleaf\\Thesis\\images\data_methods\\len_date.png')
plt.show()
```

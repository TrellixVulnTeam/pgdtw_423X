```python
import pickle
import os
from collections import defaultdict

batchData = defaultdict(list)

for filename in os.listdir("recipe421/"):
    if filename.endswith(".pickle"):
        with open("recipe421/"+filename, "rb") as infile:
            batch = pickle.load(infile)
       # print(sorted(list(batch['Units']['1']['Operations'].keys())))
        

        for key, item in batch['Units']['1']['Operations']['1.10']['PVs'].items():
            if len(item.keys()) > 1:
                batchData[int(filename[:4])].append(item)
```
